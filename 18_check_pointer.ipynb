```python
from langgraph.prebuilt import create_react_agent
from langgraph.checkpoint.memory import InMemorySaver
```

```python
def get_weather(city):
    """ Get the current weather in the city"""
    return f"The weather in {city} is sunny"
```

```python
def get_city_population(city):
    """ Get population of a given city """
    return f"The population of {city} is 5 million"
```

```python
agent = create_react_agent(
    model= 'gpt-4o',
    tools=[get_city_population, get_weather]
)
```

```python
agent.invoke(
    {
        "messages": [{"role": "user", "content": "What is the weather in sf"}]
    }
)
```

```python
agent.invoke(
    {
        "messages": [{"role": "user", "content": "What about in seattle"}]
    }
)
```

```python
checkpointer = InMemorySaver()
```

```python
agent_v1 = create_react_agent(
    model="gpt-4o",
    tools=[get_city_population, get_weather],
    checkpointer=checkpointer

)
```

```python
congif = {
    "configurable": {
        "thread_id": 1
    }
}
```

```python
agent_v1.invoke(
    {
        "messages": [{"role": "user", "content": "What is the weather in sf"}]
    },
    config=congif
)
```
